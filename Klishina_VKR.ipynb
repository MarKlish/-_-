## удаляем выбросы по правилу трех сигм
(позволяет отсечь данные, которые находятся на расстоянии больше 3-х сигм от матожидания
Для этого определяем z-оценку (по сути - стандартизацию) + для наглядности строим z-гистограмму - для этого от текущего значения отнимаем матожидание (mean) и делим получившуюся разницу на разброс (std)), затем сравниваем с 3 (т.к. "правило 3-х сигм", можно сравнивать и с 2 или с 4) и оставляем в датасете только то, что укладывается по модулю в 3 сигмы)
Альтернатива - удалять выбросы на основе межквартильного расстояния

```python
data["Матрица-наполнитель_z"] = (data["Соотношение матрица-наполнитель"] - data["Соотношение матрица-наполнитель"].mean()) / data["Соотношение матрица-наполнитель"].std()
plt.hist(data["Матрица-наполнитель_z"])
```

## **Общие подходы к оценке качества моделей**

Для **задач классификации** можно использовать:

Confusion Matrix, 

Accuracy Metric (базовая, измеряет количество верно классифицированных объектов относительно общего количества всех объектов);

Recall/Sensitivity Metric (определяет, сколько объектов наша модель смогла правильно классифицировать с позитивной меткой из всего множества позитивных); 

Precision Metric (сколько из всех объектов, которые классифицируются как положительные, действительно являются положительными, относительно общего количества полученных от модели позитивных меток), 

F1 score (сочетание precision и recall, дает некоторый компромисс между ними двумя).

Для **задач регрессии** (это текущая задача) можно использовать:

**Mean Absolute Error (MAE)** - измеряет среднюю сумму абсолютной разницы между фактическим значением и прогнозируемым значением, импортируется из sklearn.metrics как mean_absolute_error

**Mean Squared Error (MSE)** - измеряет среднюю сумму квадратной разности между фактическим значением и прогнозируемым значением, импортируется из sklearn.metrics как mean_squared_error

**Root Mean Squared Error (RMSE)** - квадратный корень от квадрата ошибки, результат легче интерпретировать, поскольку он имеет те же единицы, что и исходные значения (в отличие от MSE) и при этом на результат не накладываются отрицательные значения

**R-squared (коэффициент детерминации)** - показывает, насколько близко предсказанное значение к исходному. Диапазон от 0 до 1. Чем значение выше, тем результат работы модели лучше. Импортируется из sklearn.metrics как r2_score

В числе подбираемых параметров:

n_estimators — число «деревьев» в «случайном лесу»

max_depth — максимальная глубина деревьев

min_samples_split — минимальное число объектов, необходимое для того, чтобы узел дерева мог бы расщепиться

min_samples_leaf — минимальное число объектов в листьях

bootstrap — использование для построения деревьев подвыборки с возвращением
